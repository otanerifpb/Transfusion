```python
import pandas as pd

url = "https://raw.githubusercontent.com/otanerifpb/dataset/main/transfusion.data"

df = pd.read_csv(url, header=None)
columns = len(df.columns)

df
```

A coluna 0 (zero) é o TARGET.

```python
y = df[0]
X = df.drop(0, axis=1)

print(y)
df
```

Observe que a magnitude dos atributos é bem diferente, então cada atributo deve ser normalizado (código no mini tutorial de pandas).

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

df = pd.DataFrame(df, columns=[1,2,3,4])

df[1] = MinMaxScaler().fit_transform(np.array(df[1]).reshape(-1,1))
df[2] = MinMaxScaler().fit_transform(np.array(df[2]).reshape(-1,1))
df[3] = MinMaxScaler().fit_transform(np.array(df[3]).reshape(-1,1))
df[4] = MinMaxScaler().fit_transform(np.array(df[4]).reshape(-1,1))

df
```

Separar os dados de treinamento e testes

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = None, stratify = y)
```

Treinar 2 árvores de decisão com características diferentes

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

treino1 = DecisionTreeClassifier(criterion = "entropy")
treino2 = DecisionTreeClassifier(max_depth = 2)

treino1.fit(X_train, y_train)
treino2.fit(X_train, y_train)
```

Treinar 2 kNN com características diferentes

```python
treino3 = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean', algorithm = 'brute')
treino4 = KNeighborsClassifier(n_neighbors = 7, metric = 'manhattan', algorithm = 'brute')

treino3.fit(X_train, y_train)
treino4.fit(X_train, y_train)
```

```python
from sklearn import metrics

result1 = treino1.predict(X_test)
acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("Árvore Decisão Profundidade livre: " + "{}%".format(show1))

result2 = treino2.predict(X_test)
acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("Árvore Decisão Profundidade 2: " + "{}%".format(show2))

result3 = treino3.predict(X_test)
acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("Protocolo KNN Metric 'euclidean': " + "{}%".format(show3))

result4 = treino4.predict(X_test)
acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("Protocolo KNN Metric 'manhattan': " + "{}%".format(show4))
```
